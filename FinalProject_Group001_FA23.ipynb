Unsheltered homelessness is very common throughout California, especially in it's biggest cities like Los Angeles, San Francisco and San Diego. ## Set Up

```python
# imports
import pandas as pd

# helpers
def fetchMeltMerge(csv, id, var, value, data):
  newData = pd.read_csv(csv, skipinitialspace=True)
  newMelt = pd.melt(newData, id_vars=[id], var_name=var, value_name=value)
  
  if data.empty:
    return newMelt
  else:
    return pd.merge(data, newMelt, on=[id, var], how='outer')
```

## Data Cleaning

We have 4 `.csv` in total to parse in order to create a combined dataframe with: 
- unsheltered homeless counts (UHC)
- minimum / maximum / average temperatures
- precipitation
- year
- month

Fortunately, we've already aligned the timeframes (Jan 2012 - Dec 2022), so all we need to do is convert to a long format and unionize the variables.

```python
# dataName, dataLocation
dataFiles = [
  ('uhc', 'adata/dt_sd_unsheltered_count.csv'),
  ('minTemp', 'adata/sd_lowest_temp_92101.csv'), 
  ('avgTemp', 'adata/sd_avg_temp_92101.csv'),
  ('maxTemp', 'adata/sd_highest_temp_92101.csv'),
  ('precip', 'adata/sd_precip_92101.csv')
]

# create an empty frame to populate
df = pd.DataFrame()

# pull in all the data
for dataName, dataLocation in dataFiles:
  df = fetchMeltMerge(dataLocation, 'Year', 'Month', dataName, df)

df.head()
```

It's naturally arranged by year, and the months are strings which won't sort well. Let's convert them to numbers and sort by year-month.

```python
monthToNumber = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

df['Month'].replace(monthToNumber, inplace=True)
df = df.sort_values(by=['Year', 'Month']).reset_index(drop=True)

df.head()
```

Last thing is to make our column names consistently camelCased.

```python
df.rename(columns={'Month':'month', 'Year':'year'}, inplace=True)

df.head()
```

And voila, tidy dataframe achieved.

### Unsheltered Homeless Count and Temperature

Now, lets plot the unsheltered homeless count vs the average temperature in San Diego.

```python
sns.scatterplot(data=df, x='avgTemp', y='uhc')
plt.xlabel('Average Temperature')
plt.ylabel('Homeless Count')
plt.title('Scatter Plot of Homeless Count vs. Average Temperature')
plt.show()
```

Temperature over the full range doesn't seem to have any correlation, so lets look at individual months instead.

### Unsheltered Homeless Count and Precipitation

```python
sns.scatterplot(data=df, x='precip', y='uhc')
plt.title('Inches of Precipation vs. Unsheltered Homeless Count (UHC)')
plt.xlabel('Inches of Precipation')
plt.ylabel('Unsheltered Homeless Count')
plt.show()
```

This isn't very useful since there are so many periods with minimal rain, especially over the summer. Let's instead use a rolling value for precipitation.

```python
# calculate rolling sum over 3 month intervals
df['roll3precip'] = df['precip'].rolling(3).sum()
df.head(3)
```

A histogram will help us get a feel for the distribution of precipitation. Should see a left-leaning spread, again because of the infrequent rain.

```python
sns.histplot(data=df, x='roll3precip')
plt.xlabel("Rolled Precipation in Inches")
plt.show()
```

```python
df['roll3precip'].describe()
```

Over half of the rolling values are well under 2 inches. This makes sense with the dry months we observed in the raw data. Given how little correlation we saw there, and how many dry months we have, let's set a threshold at the top quartile of precipitation. Further, lets only look at pre-COVID, and see how it fares as a predictor of `uhc`.

```python
highRainPreCovid = df[(df['roll3precip'] > 3) & (df['year'] < 2020)]

plt.title('3 month rolling precip vs uhc')
sns.scatterplot(data=highRainPreCovid, x='roll3precip', y='uhc')
plt.show()
```

We must first acknowledge that our observation set is very small, at 26. *However*, there does appear to be some slight linearity, so let's run a regression.

```python
ols('uhc ~ roll3precip', highRainPreCovid)
```

This is encouraging, although not statistically significant based on the `P>|t| = 0.069` (above our goal of `0.05`). Can we refine the result? Let's try using the rolling `uhc` value in addition to the rolling precipitation.

```python
ols('roll3uhc ~ roll3precip', highRainPreCovid)
```

This is significant! The r-squared is `0.209` with the `P>|t| = 0.011`, though this is again marred by the limitations of our data's size.

## Attempt at Mulitvariate Analysis

Now that we have somewhat of a correlation between the 3-month rolled precipation and the 3-month rolled unsheltered homeless count for all months pre-covid. Let's make sure that the temperature in this new shortened dataframe maintains its non-linearity with respect to the 3-month rolled unsheltered homeless count.


```python
xMin = highRainPreCovid['minTemp']
xMax = highRainPreCovid['maxTemp']
xAvg = highRainPreCovid['avgTemp']
yUHC = highRainPreCovid['roll3uhc']

# Plotting the scatter plots
plt.scatter(xMin, yUHC, label='Min. Temp')
plt.scatter(xMax, yUHC, label='Max. Temp')
plt.scatter(xAvg, yUHC, label='Avg. Temp')

# Adding labels and title
plt.xlabel('Pre-Covid Max, Min, and Avg Temp')
plt.ylabel('Pre-Covid Rolled UHC')
plt.title('Pre-Covid Max, Min, and Avg Temp vs. Pre-covid Rolled UHC')
plt.legend()

# Show plot
plt.show()
```

It seems that neither max, min, or avg temperature have a linear correlation with the Rolled UHC Pre-covid, but just to be sure let's look at the pearson coefficient.

```python
# Calculate Pearson coefficient
corrMin, _ = pearsonr(xMin, yUHC)
corrMax, _ = pearsonr(xMax, yUHC)
corrAvg, _ = pearsonr(xAvg, yUHC)

corrMin_rounded = round(corrMin, 3)
corrMax_rounded = round(corrMax, 3)
corrAvg_rounded = round(corrAvg, 3)

print(f"Pearson's correlation coefficient for min temp: {corrMin_rounded}")
print(f"Pearson's correlation coefficient for max temp: {corrMax_rounded}")
print(f"Pearson's correlation coefficient for avg temp: {corrAvg_rounded}")
```

None of these values are remotely close to 1 or -1 so we can be sure that the relationship, even with a truncated pre-covid dataset, is still not linear between temperature and rolled UHC.

Since none of the variables are linear with UHC in any capacity, it doesn't make sense to do a multivariate regression analysis.

### Ethics
It's important to be clear about and understand that homelessness is not a disease, a moral failing, or even a "problem" by itself. If it were any of those things, we would in effect be claiming that there is an absolute need to pay rent or own property. Our study of homeless population trends aims to stand on its own, without normative claims. A decrease in homeless population is not "improving homelessness," since what constitutes "improvement" is not within our scope. To that end:
1. We must maintain an awareness in our work that the populations we're examining are people, often under extreme duress. 
2. We will not speak in such a way that implies a reduced count is "good," as that impinges on the humanity of those (un)counted and ignores the mortal realities of the world we study.
3. We shall refrain from normative statements such as "from 2012-2013, the homeless situation *improved*."

### Biases
We may see bias in how counts categorize who is "homeless." For example, there is often a differentiation between "homeless" and "unsheltered", with the latter referring to living out of a car or similar, but we ought be aware that the line between the two is unlikely to be standardized in all sources. We will evaluate this differentiation source by source for consistency.

### Privacy
We will never be examining individuals since the scope of our work is limited to population counts and region, and so we do not anticipate privacy concerns.
